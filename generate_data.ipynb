```python
import pandas as pd
from datetime import datetime, timedelta
import random
# This scrip will generate data for each table in our database except fact-table


# Function to generate values for date dimension table
# Dates will be genereted for 1 year from '2023-01-01'to '2023-12-31'
def generate_date_table():
    global date_id_value
    dates_value = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
    day_value = dates_value.day
    month_value = dates_value.month
    year_value = dates_value.year
    date_id_value = [i for i in range(365)]

    # week value generating
    week_value = []
    week = 1
    for i in range(0, len(dates_value)):
        if week == 53:
            week = 1

        week_value.append(week)

        if (i+1) % 7 == 0:
            week += 1

    date_table = pd.DataFrame({'date_id': date_id_value, 'date': dates_value, 'day': day_value, 'month': month_value, 'year': year_value, 'week': week_value})
    return date_table


def generate_market_price_table():
    global price_value
    global price_id_value
    currency_name = ['Â£' for _ in range(365)]
    price_id_value = [i for i in range(365)]
    price_value = [round(random.uniform(0.30, 1.40), 2) for _ in range(365)]
    market_price_table = pd.DataFrame({'price_id': price_id_value, 'currency_name': currency_name, 'price_per_kg': price_value})
    return market_price_table


def generate_crop_yield_table():
    global quantity_cropped_value
    global waste_value
    global crop_yield_id_value
    crop_yield_id_value = [i for i in range(365)]
    quantity_cropped_value = [random.randint(0, 10000) for _ in range(365)]
    waste_value = [random.randint(0, 3000) for _ in range(365)]
    crop_yield_table = pd.DataFrame({'crop_yield_id': crop_yield_id_value, 'quantity_cropped': quantity_cropped_value, 'waste': waste_value})
    return crop_yield_table


def generate_sale_table():
    # how many kilograms sold per day
    global sale_id_value
    global quantity_sold_value
    sale_id_value = [i for i in range(365)]
    quantity_sold_value = [random.randint(0, max(0, i - j)) for i, j in zip(quantity_cropped_value, waste_value)]
    total_sales_value = [round(value * price_value[index], 2) for index, value in enumerate(quantity_sold_value)]
    sales_table = pd.DataFrame({'sale_id': sale_id_value, 'total_sales': total_sales_value, 'quantity_kg': quantity_sold_value})
    return sales_table


def generate_weather_table():
    global weather_id_value
    weather_id_value = [i for i in range(365)]
    precipitation_value = [round(random.uniform(0, 5.5), 2) for _ in range(365)]
    wind_value = [random.randint(0, 80) for _ in range(365)]
    temperature_value = [random.randint(-10, 35) for _ in range(365)]
    weather_table = pd.DataFrame({'weather_id':weather_id_value, 'precipitation_mm': precipitation_value, 'wind_kmh': wind_value, 'temperature_C':temperature_value})
    return weather_table


def genrate_fact_table():
    stock_level_value = [max(0, i - j) for i, j in zip(quantity_cropped_value, quantity_sold_value)]
    product_name_value = [random.choice(['potato', 'rice', 'onion', 'wheat', 'hop']) for _ in range(365)]
    fact_table = pd.DataFrame({'stock_level_kg': stock_level_value, 'product_name': product_name_value, 'date_id': date_id_value, 'price_id': price_id_value, 'crop_yield_id': crop_yield_id_value, 'sale_id': sale_id_value, 'weather_id': weather_id_value})
    return fact_table


time_dim_df = generate_date_table()
time_dim_df.to_csv('time_dim.csv', index=False)
market_price_dim_df = generate_market_price_table()
market_price_dim_df.to_csv('market_price_dim.csv', index=False)
crop_yield_dim_df = generate_crop_yield_table()
crop_yield_dim_df.to_csv('crop_yield_dim.csv', index=False)
sales_dim_df = generate_sale_table()
sales_dim_df.to_csv('sales_dim.csv', index=False)
weather_dim_df = generate_weather_table()
weather_dim_df.to_csv('weather_dim.csv', index=False)
fact_df = genrate_fact_table()
fact_df.to_csv('fact_agri_prod.csv')

```
